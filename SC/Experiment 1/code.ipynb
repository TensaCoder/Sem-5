```python
# Identity Activation Function
import numpy as np
import matplotlib.pyplot as plt

def linear(x):
    return x

x = np.linspace(0, 10)
plt.plot(x, linear(x))
plt.title('Activation Function : Identity')
plt.show()
```

```python
# Binary Step Activation Function

import numpy as np
import matplotlib.pyplot as plt

def BinaryStep (x, threshold):
    if x>threshold:
        return 1
    else :
        return 0

x = np.linspace(-10, 10)
y=[]
threshold = int(input ("Enter the Threshold value : "))
for i in range (len(x)):
    yPoint= BinaryStep(x[i], threshold)
    y.append(yPoint)
plt.plot(x, y)
plt.title('Activation Function : Binary Step')
plt.show()
```

```python
# Sigmoid Activation Function
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x,steepness):
    expValue= steepness * x
    return 1/(1+ np.exp(-expValue))

x = np.linspace(-10, 10)
y=[]
steepness = int(input ("Enter the steepness value : "))
for i in range (len(x)):
    yPoint= sigmoid(x[i], steepness)
    y.append(yPoint)
plt.plot(x,y)
plt.title("Activation Function : Sigmoid")
plt.show()
```

```python
# Bipolar Sigmoid Activation Function
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x,steepness):
    expValue= steepness * x
    return (1 - np.exp(-expValue))/(1+ np.exp(-expValue))

x = np.linspace(-10, 10)
y=[]
steepness = int(input ("Enter the steepness value : "))
for i in range (len(x)):
    yPoint= sigmoid(x[i], steepness)
    y.append(yPoint)
plt.plot(x,y)
plt.title("Activation Function : Bipolar Sigmoid")
plt.show()
```

```python
# ReLU Activation Function
import numpy as np
import matplotlib.pyplot as plt

def ReLU(x):
    if x< 0:
        return 0
    else:
        return x

x=np.linspace(-10, 10)
y=[]
for i in range (len(x)):
    yPoint= ReLU(x[i])
    y.append(yPoint)
plt.plot(x,y)
plt.title("Activation Function : ReLU ")
plt.show()
```
